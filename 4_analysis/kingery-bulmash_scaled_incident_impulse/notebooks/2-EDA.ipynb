```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('pub')
```

### File Locations

```python
today = datetime.today()
in_file = Path.cwd().parents[2] / "2_data" / "kingery-bulmash_scaled_incident_impulse" / "processed" / "summary_Dec-19-2018.pkl"

#report_dir = Path.cwd().parents[2] / "reports"
#report_file = report_dir / "Excel_Analysis_{today:%b-%d-%Y}.xlsx"
```

```python
df = pd.read_pickle(in_file)
```

### Perform Data Analysis

```python
x_cv = df['scaled_distance'].loc[df['Reference'] == 'kb_curve'].values
y_cv = df['scaled_incident_impulse'].loc[df['Reference'] == 'kb_curve'].values
x_02 = df['scaled_distance'].loc[df['Reference'] == 'REF 2'].values
y_02 = df['scaled_incident_impulse'].loc[df['Reference'] == 'REF 2'].values
x_67 = df['scaled_distance'].loc[df['Reference'] == 'REF 6 & 7'].values
y_67 = df['scaled_incident_impulse'].loc[df['Reference'] == 'REF 6 & 7'].values
x_08 = df['scaled_distance'].loc[df['Reference'] == 'REF 8'].values
y_08 = df['scaled_incident_impulse'].loc[df['Reference'] == 'REF 8'].values
x_09 = df['scaled_distance'].loc[df['Reference'] == 'REF 9'].values
y_09 = df['scaled_incident_impulse'].loc[df['Reference'] == 'REF 9'].values
```

```python
fig, ax = plt.subplots()
ax.loglog(x_cv, y_cv, label = 'Kingery Bulmash Curve')
ax.plot(x_02, y_02, 'd', label = 'Kingery [6]', mfc = 'none')
ax.plot(x_67, y_67, 'o', label = 'Reisler, Pettit, and Kennedy [1][3]', mfc = 'none')
ax.plot(x_08, y_08, '*', label = 'Swisdak [4]')
ax.plot(x_09, y_09, 's', label = 'Nuclear Weapons Blast Phenomena Volume IV [5]', mfc='none')
ax.legend()
ax.set_ylabel(r'Scaled Incident Impulse $\left(\frac{psi-ms}{\sqrt{lb^{1/3}}}\right)$')
ax.set_xlabel(r'Scaled Distance $\left(\frac{ft}{\sqrt{lb^{1/3}}}\right)$')
plt.show();
```
