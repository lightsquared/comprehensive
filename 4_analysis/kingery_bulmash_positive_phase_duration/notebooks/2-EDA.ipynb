```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('pub')
```

### File Locations

```python
today = datetime.today()
in_file_cv = Path.cwd().parents[2] / "2_data" / "kingery_bulmash_positive_phase_duration" / "processed" / "summary_cv_Dec-18-2018.pkl"
in_file_67 = Path.cwd().parents[2] / "2_data" / "kingery_bulmash_positive_phase_duration" / "processed" / "summary_67_Dec-18-2018.pkl"
in_file_08 = Path.cwd().parents[2] / "2_data" / "kingery_bulmash_positive_phase_duration" / "processed" / "summary_08_Dec-18-2018.pkl"
in_file_09 = Path.cwd().parents[2] / "2_data" / "kingery_bulmash_positive_phase_duration" / "processed" / "summary_09_Dec-18-2018.pkl"

#report_dir = Path.cwd().parents[2] / "reports"
#report_file = report_dir / "Excel_Analysis_{today:%b-%d-%Y}.xlsx"
```

```python
df_cv = pd.read_pickle(in_file_cv)
df_67 = pd.read_pickle(in_file_67)
df_08 = pd.read_pickle(in_file_08)
df_09 = pd.read_pickle(in_file_09)
```

### Perform Data Analysis

```python
x_cv = df_cv['scaled_distance'].values
y_cv = df_cv['time_of_arrival'].values
x_67 = df_67['scaled_distance'].values
y_67 = df_67['time_of_arrival'].values
x_08 = df_08['scaled_distance'].values
y_08 = df_08['time_of_arrival'].values
x_09 = df_09['scaled_distance'].values
y_09 = df_09['time_of_arrival'].values

fig, ax = plt.subplots()
ax.loglog(x_cv, y_cv, label = 'Kingery Bulmash Curve')
ax.plot(x_67, y_67, 'o', label = 'Reisler, Pettit, and Kennedy [1][3]', mfc = 'none')
ax.plot(x_08, y_08, '*', label = 'Swisdak [4]')
ax.plot(x_09, y_09, 's', label = 'Nuclear Weapons Blast Phenomena Volume IV [5]', mfc='none')
ax.legend()
ax.set_ylabel(r'Scaled Positive Phase Duration $\left(\frac{ms}{\sqrt{lb^{1/3}}}\right)$')
ax.set_xlabel(r'Scaled Distance $\left(\frac{ft}{\sqrt{lb^{1/3}}}\right)$')
plt.show();
```
